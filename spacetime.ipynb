```python
import datetime
import geopy.distance
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams["pgf.texsystem"] = "pdflatex"
matplotlib.rcParams["font.family"] = "None"
matplotlib.rcParams["pgf.preamble"] = "\\usepackage[utf8]{inputenc}\\DeclareUnicodeCharacter{2212}{-}"
```

# Load data set

```python
df = pd.read_feather('SRFG-v1.fth')
```

# Autocorrelation - Time

```python
df_tdd = df[["time", "datarate", "datarate"]].copy() # time datarate datarate
df_tdd.columns = ["time", "dr_pre", "dr"]

df_s = pd.DataFrame({'dr_pre':[],'dr':[]})
df_timecorr = pd.DataFrame({'seconds':[],'autocorr':[]})
for s in range(0,20):
    rowdiff = s    
    df_s = df_s[0:0]
    while rowdiff >= 0:
        df_tdd.loc[:, "dr_pre"] = df_tdd.loc[:, "dr"].shift(rowdiff)
        df_s = df_s.append(df_tdd.loc[df_tdd.time.diff(rowdiff) == datetime.timedelta(seconds = s), ["dr_pre", "dr"]])
        rowdiff -= 1
    df_timecorr = df_timecorr.append({'seconds':s, 'autocorr':df_s.corr().iloc[1,0]}, ignore_index=True)
plt.plot(df_timecorr["autocorr"])

df_tdd = df[["time", "rssi", "rssi"]].copy() # time rssi rssi
df_tdd.columns = ["time", "dr_pre", "dr"]

df_s = pd.DataFrame({'dr_pre':[],'dr':[]})
df_timecorr = pd.DataFrame({'seconds':[],'autocorr':[]})
for s in range(0,20):
    rowdiff = s    
    df_s = df_s[0:0]
    while rowdiff >= 0:
        df_tdd.loc[:, "dr_pre"] = df_tdd.loc[:, "dr"].shift(rowdiff)
        df_s = df_s.append(df_tdd.loc[df_tdd.time.diff(rowdiff) == datetime.timedelta(seconds = s), ["dr_pre", "dr"]])
        rowdiff -= 1
    df_timecorr = df_timecorr.append({'seconds':s, 'autocorr':df_s.corr().iloc[1,0]}, ignore_index=True)
plt.plot(df_timecorr["autocorr"],'--')

plt.xlabel('Time difference of measurements [s]')
plt.ylabel('Correlation coefficient')
plt.grid()
plt.gcf().set_size_inches(5, 3)
plt.savefig('figures/autocorrelation_time_20s.pgf',bbox_inches='tight');
```

# Autocorrelation - Space

```python
distance = df["dlong"].abs().mean() # set distance (in longitude)
intervalWidth = distance / 2        # set width of interval (in longitude, one side)

df_tldd = df[["time", "long", "datarate",  "datarate"]].copy() # tlbb = time long bandwidth bandwidth 
df_tldd.columns = ["time", "long", "bwd", "bwd_pre"]
df_distance = pd.DataFrame({'bwd_pre':[],'bwd':[]})
df_spacecorr = pd.DataFrame({'distance':[],'autocorr':[]}) 
for s in range(0,60):
    rowdiff = 0
    df_distance = df_distance[0:0]
    while rowdiff <= s + 90:
        df_tldd["bwd_pre"] = df_tldd["bwd"].shift(rowdiff)
        df_distAndTimeOK = df_tldd[((abs(df_tldd.long.diff(rowdiff)) < s * distance + intervalWidth) & 
            (abs(df_tldd.long.diff(rowdiff)) > s * distance - intervalWidth)) & 
            (abs(df_tldd.time.shift(rowdiff) - df_tldd.time) < datetime.timedelta(hours = 0.75))]
        df_distance = df_distance.append(df_distAndTimeOK.loc[:,("bwd_pre", "bwd")])
        rowdiff +=1
    df_spacecorr = df_spacecorr.append({'distance':s * distance, 'autocorr':df_distance.corr().iloc[1,0]}, ignore_index=True)

df_spacecorr["distancem"] = df_spacecorr["distance"] * geopy.distance.geodesic((df["lat"].mean(), df["long"].mean()), (df["lat"].mean(), df["long"].mean()+1)).meters
plt.plot(df_spacecorr["distancem"],df_spacecorr["autocorr"])

distance = df["dlong"].abs().mean() # set distance (in longitude)
intervalWidth = distance / 2        # set width of interval (in longitude, one side)

df_tldd = df[["time", "long", "rssi",  "rssi"]].copy() # tlbb = time long rssi rssi 
df_tldd.columns = ["time", "long", "bwd", "bwd_pre"]
df_distance = pd.DataFrame({'bwd_pre':[],'bwd':[]})
df_spacecorr = pd.DataFrame({'distance':[],'autocorr':[]}) 
for s in range(0,60):
    rowdiff = 0
    df_distance = df_distance[0:0]
    while rowdiff <= s + 90:
        df_tldd["bwd_pre"] = df_tldd["bwd"].shift(rowdiff)
        df_distAndTimeOK = df_tldd[((abs(df_tldd.long.diff(rowdiff)) < s * distance + intervalWidth) & 
            (abs(df_tldd.long.diff(rowdiff)) > s * distance - intervalWidth)) & 
            (abs(df_tldd.time.shift(rowdiff) - df_tldd.time) < datetime.timedelta(hours = 0.75))]
        df_distance = df_distance.append(df_distAndTimeOK.loc[:,("bwd_pre", "bwd")])
        rowdiff +=1
    df_spacecorr = df_spacecorr.append({'distance':s * distance, 'autocorr':df_distance.corr().iloc[1,0]}, ignore_index=True)

df_spacecorr["distancem"] = df_spacecorr["distance"] * geopy.distance.geodesic((df["lat"].mean(), df["long"].mean()), (df["lat"].mean(), df["long"].mean()+1)).meters
plt.plot(df_spacecorr["distancem"],df_spacecorr["autocorr"],'--')

plt.xlabel('Space difference of measurements [m]')
plt.ylabel('Correlation coefficient')
plt.grid()
plt.gcf().set_size_inches(5, 3)
plt.savefig('figures/autocorrelation_space.pgf',bbox_inches='tight');
```

# Autocorrelation - Time and Space

```python
distance = df["dlong"].abs().mean()    # set distance (in longitude)
intervalWidth = distance / 2 
min_samples = 100

df_tldd = df[["time", "long", "long", "datarate",  "datarate"]].copy() 
df_tldd.columns = ["time", "long", "long_pre", "dr", "dr_pre"]

df_timeOK = pd.DataFrame()
df_timespace = pd.DataFrame({"seconds":[], "distance":[], "corr":[], "n":[]})

for time in range(60):
    rowdiff = time    
    df_timeOK = df_timeOK[0:0]
    while rowdiff >= 0:
        df_tldd["dr_pre"] = df_tldd["dr"].shift(rowdiff)
        df_tldd["long_pre"] = df_tldd["long"].shift(rowdiff)
        df_timeOK = df_timeOK.append(df_tldd[df_tldd.time.diff(rowdiff) == datetime.timedelta(seconds = time)])
        rowdiff -= 1
    for space in range(60):
        df_distOK = df_timeOK[((abs(df_timeOK.long_pre - df_timeOK.long)) < space * distance + intervalWidth) & 
            ((abs(df_timeOK.long_pre - df_timeOK.long)) > space * distance - intervalWidth)]
        df_timespace = df_timespace.append({"seconds": time, 
            "distance": space * distance, 
            "corr": df_distOK.corr().loc["dr","dr_pre"] ,
            "n": df_distOK.time.count()}, ignore_index = True)
df_timespace.loc[df_timespace.n < min_samples, 'corr'] = np.nan

df_timespace["distancem"] = df_timespace["distance"] * geopy.distance.geodesic((df["lat"].mean(), df["long"].mean()), (df["lat"].mean(), df["long"].mean()+1)).meters

df_timespace_wide = df_timespace.pivot_table(index='distancem', columns='seconds', values='corr')
ax = plt.imshow(df_timespace_wide, origin='lower', interpolation='none', cmap='cividis')
ax.axes.set_yticks([y/distance/geopy.distance.geodesic((df["lat"].mean(), df["long"].mean()), (df["lat"].mean(), df["long"].mean()+1)).meters for y in range(0,1501,250)])
ax.axes.set_yticklabels(range(0,1501,250));
plt.xlabel('Time difference of measurements [s]')
plt.ylabel('Space difference of measurements [m]')
plt.clim(-1,1)
plt.colorbar()
plt.gcf().set_size_inches(5, 3)
plt.savefig('figures/autocorrelation_timespace.pgf',bbox_inches='tight');
```
